```python
import pandas
```

```python
rhein = pandas.read_csv("100046-Rhein2023.csv", sep=";")
rhein
```

```python
rhein['minute']  = rhein['Startzeitpunkt'].str.slice(0,16).astype(str) ## variable
rhein
```

```python
rhein_kurz = rhein.loc[:, ['Temperatur [°C]', 'minute']]
print(rhein_kurz)
```

```python
rosental = pandas.read_csv('100294-Rosental.csv', sep=";")
rosental
```

```python
rosental['minute'] = rosental['Zeitstempel'].str.slice(0, 16).astype(str)
rosental
```

```python
rosental_kurz = rosental.loc[:, ['minute', 'Lufttemperatur']]
print(rosental_kurz)
```

```python
rosental_kurz['minute']
```

```python
rhein_kurz['minute']
```

```python
merged = pandas.merge(rhein_kurz, rosental_kurz, on="minute", left_index=False, right_index=False)
res = merged.sort_values('minute')
res['diff'] = res['Temperatur [°C]'] - res['Lufttemperatur']
res['Datum'] = res['minute'].str.slice(0, 10)
res['Tageszeit'] =res['minute'].str.slice(11, 16)
res = res.loc[:, ['minute', 'Datum', 'Tageszeit', 'Temperatur [°C]', 'Lufttemperatur', 'diff']]
res = res.rename(columns={'Temperatur [°C]':'Wassertemperatur',
           'diff': 'WLdiff'}, copy=True)
print(res)
```

```python
res.to_csv('2023-06-23-merged.tsv', sep="\t", index=False)
```
