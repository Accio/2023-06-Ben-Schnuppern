```python
import pandas
```

```python
rhein = pandas.read_csv("100046-Rhein2023.csv", sep=";")
rhein
```

```python
rhein['minute']  = rhein['Startzeitpunkt'].str.slice(0,16).astype(str) ## variable
rhein
```

```python
rhein_kurz = rhein.loc[:, ['Temperatur [°C]', 'minute']]
print(rhein_kurz)
```

```python
rosental = pandas.read_csv('100294-Rosental.csv', sep=";")
rosental
```

```python
rosental['minute'] = rosental['Zeitstempel'].str.slice(0, 16).astype(str)
rosental
```

```python
rosental_kurz = rosental.loc[:, ['minute', 'Lufttemperatur']]
print(rosental_kurz)
```

```python
rosental_kurz['minute']
```

```python
rhein_kurz['minute']
```

```python
merged = pandas.merge(rhein_kurz, rosental_kurz, on="minute", left_index=False, right_index=False)
res = merged.sort_values('minute')
res['Zeit']=pandas.to_datetime(res['minute'], format="%Y-%m-%dT%H:%M")
res['diff'] = round(res['Temperatur [°C]'] - res['Lufttemperatur'], 2)
res['Datum'] = pandas.to_datetime(res['minute'].str.slice(0, 10))
res['Tageszeit'] =res['minute'].str.slice(11, 16)
res = res.loc[:, ['minute', 'Zeit', 'Datum', 'Tageszeit', 'Temperatur [°C]', 'Lufttemperatur', 'diff']]
res = res.rename(columns={'Temperatur [°C]':'Wassertemperatur',
           'diff': 'WLdiff'}, copy=True)
print(res)
```

```python
ax = res.plot(x="Zeit", y=['Wassertemperatur', 'Lufttemperatur', 'WLdiff'],
              xlabel="Zeit", ylabel="Temperatur (Celsius)",
              kind="line")
ax.axhline(y=0, linestyle="dotted", color="red")
```

```python
ax = res.boxplot(column='WLdiff', by="Tageszeit", figsize=(10,6), rot=45,
                 xlabel="Tageszeit", ylabel="Unterschied (Celsius)", grid=False)
ax.axhline(y=0, linestyle="dotted", color="red")
```

We also report aggreated statistics

```python
import numpy as np

res_agg = res.iloc[:, ~res.columns.isin(['minute', 'Zeit', 'Tageszeit'])].groupby('Datum').agg([min, max, np.ptp, np.median,np.std])
res_agg
```
